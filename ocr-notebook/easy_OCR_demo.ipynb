```python
import easyocr
from PIL import Image
import matplotlib.pyplot as plt

# Image path
img_path = r"C:\number_pate_reco\plates\scaned_img_0.jpg"

# Show the image
Image.open(img_path).show()

# EasyOCR Reader
reader = easyocr.Reader(['en'])

# Detect text
result = reader.readtext(img_path)
print("OCR Result:", result)

# Safe coordinate extraction
if result:
    cord = result[0][0]  # first detected bounding box
    if cord and all(len(point) == 2 for point in cord):
        x_min = min(point[0] for point in cord)
        y_min = min(point[1] for point in cord)
        x_max = max(point[0] for point in cord)
        y_max = max(point[1] for point in cord)
        print("Coordinates:", x_min, y_min, x_max, y_max)
    else:
        print("Error: Bounding box format invalid!")
else:
    print("No text detected! Check the image quality or path.")
```
